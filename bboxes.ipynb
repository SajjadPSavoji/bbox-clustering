```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from PIL import Image

from anchor import avg_iou , kmeans
```

```python
def load_data(csv_path):
    df = pd.read_csv(csv_path)
    w = (df.w2-df.w1)/df.W
    h = (df.h2-df.h1)/df.H
    data_ = []
    data_.append(np.array(w))
    data_.append(np.array(h))
    data_ = np.array(data_).reshape(-1, 2)
    return data_
```

```python
data = load_data('data/bboxes.csv')
```

```python
plt.scatter(data[:, 0], data[:, 1], alpha=0.1)
plt.grid()
plt.xlabel('w')
plt.ylabel('h')
```

```python
out, labels = kmeans(data, k=5)
print("Accuracy: {:.2f}%".format(avg_iou(data, out) * 100))
print("Boxes:\n {}".format(out))

ratios = np.around(out[:, 0] / out[:, 1], decimals=2).tolist()
print("Ratios:\n {}".format(sorted(ratios)))
```

```python
plt.scatter(data[:, 0], data[:, 1], alpha=0.1, c=labels)
plt.scatter(out[:, 0], out[:, 1], c='black')
plt.grid()
plt.xlabel('w')
plt.ylabel('h')
```

```python
out, labels = kmeans(data, k=9)
print("Accuracy: {:.2f}%".format(avg_iou(data, out) * 100))
print("Boxes:\n {}".format(out))

ratios = np.around(out[:, 0] / out[:, 1], decimals=2).tolist()
print("Ratios:\n {}".format(sorted(ratios)))
```

```python
plt.scatter(data[:, 0], data[:, 1], alpha=0.1, c=labels)
plt.scatter(out[:, 0], out[:, 1], c='black')
plt.grid()
plt.xlabel('w')
plt.ylabel('h')
```

```python
clt = KMeans(n_clusters=5)
```

```python
clt.fit(data)
```

```python
labels = clt.predict(data)
out = clt.cluster_centers_
```

```python
plt.scatter(data[:, 0], data[:, 1], alpha=0.1, c=labels)
plt.scatter(out[:, 0], out[:, 1], c='black')
plt.grid()
plt.xlabel('w')
plt.ylabel('h')
```

```python
clt = KMeans(n_clusters=9)
```

```python
clt.fit(data)
labels = clt.predict(data)
out = clt.cluster_centers_
```

```python
plt.scatter(data[:, 0], data[:, 1], alpha=0.1, c=labels)
plt.scatter(out[:, 0], out[:, 1], c='black')
plt.grid()
plt.xlabel('w')
plt.ylabel('h')
```
